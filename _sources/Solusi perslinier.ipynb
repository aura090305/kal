# **Solusi Persamaan Linier**

Solusi persamaan linier adalah nilai dari variabel yang memenuhi persamaan tersebut. Jika ada lebih dari satu persamaan dalam sistem, solusi harus memenuhi semua persamaan secara bersamaan.

\
a. Solusi Unik (Tunggal)

- Terdapat satu pasangan nilai variabel yang memenuhi semua persamaan.
- Grafisnya, dua garis dalam sistem berpotongan di satu titik.
- Syarat: Koefisien variabel harus memiliki determinan ≠ 0 (tidak sejajar atau identik)

contoh:

\begin{array}{cc}
2x+3y=6 \\
x-y=1 \\
\end{array}


solusi: (x,y)=(3,2) (setelah diselesaikan dengan metode substitusi/eliminasi).

\
b. Solusi Tak Hingga (Tak Terbatas)

- Terdapat banyak solusi yang memenuhi persamaan.
- Grafisnya, kedua garis berimpit (identik).
- Syarat: Dua persamaan setara, hanya berbeda skala.

contoh:

\begin{array}{cc}
2x+3y=6 \\
4x+6y=12 \\
\end{array}

Solusi: Tak hingga banyaknya, karena persamaan kedua merupakan kelipatan pertama.

\
c. Tidak Ada Solusi (Inkosisten)

- Tidak ada nilai variabel yang memenuhi kedua persamaan.
- Grafisnya, dua garis sejajar dan tidak berpotongan.
- Syarat: Koefisien variabel sama, tetapi konstanta berbeda.

contoh:

\begin{array}{cc}
2x+3y=6 \\
4x+6y=10 \\
\end{array}

Solusi: Tidak ada solusi, karena kedua garis sejajar tetapi tidak identik.



# **Metode Eliminasi**

Metode Eliminasi adalah teknik dasar dalam menyelesaikan sistem persamaan linier dengan cara menghilangkan (mengeliminasi) satu variabel dengan menjumlahkan atau mengurangkan persamaan setelah mengalikan salah satu atau kedua persamaan agar memiliki koefisien yang sama.

\
Langkah-Langkah Metode Eliminasi

1️⃣ Samakan koefisien salah satu variabel (jika perlu, kalikan salah satu atau kedua persamaan).

2️⃣ Kurangkan atau jumlahkan persamaan untuk menghilangkan variabel tersebut.

3️⃣ Selesaikan persamaan satu variabel yang tersisa.

4️⃣ Substitusikan nilai yang diperoleh ke persamaan awal untuk menemukan nilai variabel lain.

- **Contoh**

Diberikan persamaan linier:

\begin{array}{cc}
2x+3y=8 \\
4x-2y=10 \\
\end{array}

\
**Langkah 1: Samakan koefisien salah satu variabel**

- Kita samakan koefisien X dengan mengalikan persamaan pertama dengan 2, sehingga menjadi:

\begin{array}{cc}
4x+6y=16 \\
\end{array}

\
sistem baru menjadi:

\begin{array}{cc}
4x+6y=16 \\
4x-2y= 10 \\
\end{array}

\
**Langkah 2: Kurangkan kedua persamaan**

\begin{array}{cc}
(4x+6y)-(4x-2y) = 10-16 \\
4x+6y-4x+2y= 6 \\
8y=6
y=\frac{6}{8}=\frac{3}{4}
\end{array}

\
**Langkah 3: Substitusi nilai y ke persamaan pertama**

\begin{array}{cc}
2x+3 (\frac{3}{4})=8 \\
2x+ \frac{9}{4}=8 \\
2x=8- \frac{9}{4} \\
2x= \frac{32}{4}- \frac{9}{4}= \frac{23}{4} \\
x=\frac{23}{8}
\end{array}

\
jadi:
\begin{array}{cc}
X= \frac{23}{8} \\
y= \frac{3}{4}
\end{array}



# **Metode Eliminasi Gauss**

Metode Eliminasi Gauss adalah teknik penyelesaian sistem persamaan linier menggunakan matriks augmented dan Operasi Baris Elementer `(OBE)` untuk mengubahnya menjadi bentuk eselon baris atas `(upper triangular form)`.

\
**Langkah-Langkah Metode Eliminasi**

1️⃣ Samakan koefisien salah satu variabel (jika perlu, kalikan salah satu atau kedua persamaan).

2️⃣ Kurangkan atau jumlahkan persamaan untuk menghilangkan variabel tersebut.

3️⃣ Selesaikan persamaan satu variabel yang tersisa.

4️⃣ Substitusikan nilai yang diperoleh ke persamaan awal untuk menemukan nilai variabel lain.
